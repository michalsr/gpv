```python
edge_dict = {}
neighbor_freq = {}
neighbors_seen = {}
for c in unseen:
    if c in dict_list['cat_to_num'].keys():
        total = 0
        
        num = dict_list['cat_to_num'][c]
        
        for n in net[num]:
           # if net.nodes[n]['title'] == 'red wine glass':
#                 print(len(net[n]))
            for y in net[n]:
                if y not in neighbors_seen:
                    neighbors_seen[y] = 1
                else:
                    neighbors_seen[y] += 1
            neighbor_freq[net.nodes[n]['title']] = len(net[n])
            total+= len(net[n])
            edge_dict[c] = total
      
    else:
        print(c)

print(edge_dict)
    
```

```python
top_ones = []
sorted_values = sorted(neighbor_freq.values())
for x in reversed(sorted_values):
    for k in neighbor_freq:
        if neighbor_freq[k] == x:
            top_ones.append((k,x))
print(top_ones)
```
